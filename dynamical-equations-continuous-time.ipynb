```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
for P0 in np.arange(0, 0.5, 0.1):
    for E0 in np.arange(0, 0.5, 0.1):
        P, E, t = P0, E0, 0
        a, b, K = 1, 1, 1
        Dt = 0.0001
        Pdata, Edata, tdata = [P], [E], [t]
        while t < 10:
            nP = P + (a*P*(1-P/E))*Dt
            nE = E + (b*E*(1-E/(K-P)))*Dt
            t = t + Dt
            P, E = nP, nE
            Pdata.append(P)
            Edata.append(E)
            tdata.append(t)
        #plt.plot(tdata, Pdata);
        #plt.plot(tdata, Edata);
        plt.plot(Pdata, Edata);
```
