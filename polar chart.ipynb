```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data.csv')
```

```python
df.head()
```

```python
dec = df.Distance
ra = df.Radial
n=range(100)

#plt.figure(figsize = (15,10))
ra = [x/180.0*3.141593 for x in ra]
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1.1,1.1],polar=True)
ax.set_ylim(0,60)
ax.set_yticks(np.arange(0,60,10))
ax.set_rlabel_position(0)
#colors=['k' if a < 5000 else 'b' if a < 20000 else 'g' for a in n]
sc=ax.scatter(ra,dec,c = n)
ax.set_theta_zero_location('N', offset=0)
ax.set_title("Where did Collisions Happen?", va='bottom',y=1.1)
cax = fig.add_axes([1.2,0.1,0.03,1])
fig.colorbar(sc, cax=cax, label="colorbar / Altitude", ticks=[10,50,80])
plt.subplots_adjust(top=0.8)
plt.savefig('polarchart.png')
plt.show()
```
