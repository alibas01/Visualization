```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
xdata = np.random.rand(100)
ydata = np.random.rand(100)
```

```python
plt.plot(xdata, ydata, 'ko')
plt.title('A random scatter plot')
plt.xlabel('$x$')
plt.ylabel('$y$')
#plt.show()
```

```python
plt.hist(xdata, bins = 100);
```

```python
x = np.linspace(-200, 200, 10000)
plt.plot(x, np.sin(x))
plt.xlabel('$x$')
plt.ylabel('$\sin(x)$')
plt.xlim(-10, 10)
plt.ylim(-2, 2)
```

```python
plt.pie(np.random.rand(10));
```
